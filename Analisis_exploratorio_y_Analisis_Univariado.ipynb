1) ANALISIS UNIVARIADO

Este analisis implica analisar una variable a la vez, podemos usar histogramas, diagramas de caja y graficos de densidad para observar la distribución de una variable, esto nos da información acerca de como es la variacion y comportamiento de esta variable.

GRÁFICO #1

```python
#Construcion del histograma de "total_bill"
sns.histplot(data=df, x='total_bill')
plt.plot()
```

- Este histograma nos genero una distribución de la cuenta total de las cuentas que hay, asi que, observamos aqui que la mayor frecuencia en las cuentas a pagar recae entre los 10 y 25 dolares

GRÁFICO #2

```python
sns.boxenplot(data=df, x='total_bill')
plt.plot()
```

- Aqui observamos que en esta gráfica de boxplot se ve de una manera mas clara la frecuencia entre el rango de 10 y 25 donde encontramos la mayoria de datos, asi como podemos observar los valores atipicos estos estan fuera de nuestro analisis

GRÁFICO #3

```python
sns.kdeplot(data=df, x='total_bill')
plt.show()
```

- En este tercer gráfico podemos ver la densidad y total_bill, asi como en los dos gráficos anteriores se muestra el rango donde mas hay frecuencia

- Estos 3 gráficos son muy utiles en el Analisis Univariado

- CONCLUSIÓN: En base al analisis hecho con el dataset llamado "Tips" perteneciente a Seaborn podemos destacar lo siguiente:

  Analisamos la variables "total_bill" la cual nos dice los siguiente:

  En el "gráfico 1" nos muestra que el rango de dolares que se pagan en cuentas recae entre los 10 y 20 dolares, asi mismo complementamos entre analisis con el "gráfico 2"  y "gráfico 3" donde visualizamos de una manera más clara este frecuencia de los datos y observamos tambien los valores atipicos que se presentan en el analisis.


2) ANALISIS BIVARIADO

*   CONCLUSIÓN: En el "gráfico 1" pairplot se comparo la variable "sex" donde se puede concluir lo siguiente:

1) Con respecto a la variable "total_bill" y la variable "tip" se oberva que el sexo masculino tiende a dar una mayor porpina con una frecuencia de 2 a 5 dolares

2) Con respecto a la variable "total_bil" y la variable "size" se oberva una mayor frecuencia en el sexo masculino el cual suele pagar cuentas de 20 a 40 dolares con un rango de 3 a 4 personas

3) Con respecto a la variable "tip" y la variable "size" se observa que hay una mayor frecuencia de propinas entre los 3 a 6 dolares dadas por mesas de 3 a 4 personas y esta tendencia se oberva mayormente en el sexo masculino

4) Con respecto a la variable "tip" y la variable "tip" se observa que hay una mayor frecuencia de propinas dadas por el sexo masculino

5) Con respecto a la variable "size" y la variable "size" se oberva que suele haber una mayor frecuencia del sexo masculino con mesas de 2 personas que acomparacion del sexo femenino tiene una mayor frecuencia de 2 personas, es decir, el sexo masculno suele salir mas frecuentemente en pares que el femenino

4. ANALISIS EXPLORATORIO (EDA) - Exploratory Data Analysis

Se utiliza para analizar e investigar conjuntos de datos y resumir sus características principales, a menudo empleando métodos de visualización de datos. EDA ayuda a determinar la mejor manera de manipular las fuentes de datos para obtener las respuestas que necesita, lo que facilita descubrir patrones, detectar anomalías, poner a prueba una hipótesis o verificar suposiciones.

EDA se utiliza principalmente para ver qué datos se pueden revelar más allá de la tarea formal de modelado o de la prueba de hipótesis y proporciona una mejor comprensión de las variables del conjunto de datos y las relaciones entre ellas. También puede ayudar a determinar si las técnicas estadísticas que se está considerando para el análisis de datos son adecuadas.

*   Para este caso práctico usaremos el dataset de Seaborn llamado "Tips"

```python
sns.set(style='darkgrid')
df = sns.load_dataset('tips')
df
```

*   Comenzamos haciendo un gráfico de barras el cual nos muestra las propinas promedio por sexo

GRÁFICO #1

```python
#GRÁFICO DE BARRAS DE PROPINAS PROMEDIO POR SEXO
sns.barplot(data=df, x='sex', y='tip', errorbar=None)
plt.title('Propinas promedio por sexo')
plt.show()
```

*   Aquí se puede observar que en promedio el sexo masculino da más propina que el sexo femenino

GRÁFICO #2

*   Aquí se quiere demostrar si ¿La propina que se paga cada día es la misma sin importar el dia de la semana?

```python
sns.barplot(data=df, x='day', y='tip', errorbar=None)
plt.title('Propinas promedio por día')
plt.show()
```

*   Se observa que los Domingos son los días que se pagan más propinas

GRÁFICO #3

*   Ahora queremos ver las propinas promedio por tiempo

```python
sns.barplot(data=df, x='time', y='tip', errorbar=None)
plt.title('Propinas promedio por tiempo')
plt.show()
```

*   Se observa que se pagan más propinas en las cenas que en el almuerzo

GRÁFICO #4

*  Se quiere observar si que un cliente fume o no afecte a las propinas que da

```python
sns.barplot(data=df, x='smoker', y='tip', errorbar=None)
plt.title('Propinas promedio por Fumador/No fumador')
plt.show()
```

*   Se observa que el hecho de que el cliente fume o no, no termina afectando a las propinas

GRÁFICO #5

*   Se quiere mostrar el promedio de propinas por diá segun el sexo

```python
sns.barplot(data=df, x='day', y='tip', hue='sex', errorbar=None)
plt.title('Propinas promedio por diá y sexo')
plt.show()
```

*  Se oberserva que los domingos son el dia con más propina dadas por el sexo femenino

GRÁFICO #6

*  Se quiere saber las propinas promedio por tiempo y sexo


```python
sns.barplot(data=df, x='time', y='tip', hue='sex', errorbar=None)
plt.title('Propinas promedio por tiempo y sexo')
plt.show()
```

*  Se observa que las mejores propinas se dan en la cena por el sexo masculino

GRÁFICO #7

* Se quiere saber el Promedio de propinas por fumador/No fumador y sexo

```python
sns.barplot(data=df, x='smoker', y='tip', hue='sex', errorbar=None)
plt.title('Propinas promedio por Fumador/No fumador y sexo')
plt.show()
```

* Se observa que las mejores propinas son dadas por No Fumadores masculinos

* Crearemos una nueva columna en nuestro DataFrame que calcule el procentaje de propina

```python
df['tip_percentage'] = (df['tip']/df['total_bill']) * 100
df
```

* Haremos un gráfico de barras de porcentaje de propina promedio por sexo

GRÁFICO #10

```python
sns.barplot(data=df, x='sex', y='tip_percentage', errorbar=None)
plt.title('Porcenta de propina promedio por sexo')
plt.show()
```

* Se observa que las mujeres dan mas % del total de sus cuentas,es decir, compran menos

GRÁFICO #11

* Queremos observar mediante un gráfico de barras el procentaje de propina promedio por día

```python
sns.barplot(data=df, x='day', y='tip_percentage', errorbar=None)
plt.title('Porcenta de propina promedio por dia')
plt.show()
```

* Se observa que los Viernes hay un mayor porcentaje de propinas pero sin una diferencia significativa de los diás Jueves y Domingos asi como el día más bajo es el Sabado

GRÁFICO #12

* Queremos observar el porcentaje de propina promedio tiempo

```python
sns.barplot(data=df, x='time', y='tip_percentage', errorbar=None)
plt.title('Porcentaje de propina promedio por tiempo')
plt.show()
```

* Se puede observar que no hay una diferencia significativa pero en el almuerzo suele darse mas propinas

GRÁFICO #13

* Se quiere realizar un gráfico del porcentaje promedio de propina por fumador/No fumador

```python
sns.barplot(data=df, x='smoker', y='tip_percentage', errorbar=None)
plt.title('Porcentaje de propina promedio por Fumador/No fumador')
plt.show()
```

* Nuevamente vemos que no hay una diferencia significativa, es decir, no importa si la persona fuma o no te dara la misma propia
